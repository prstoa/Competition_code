```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

* 장·단기, 환불
  - 장기 : 한달 결재 횟수가 많고(이상치) 한번 지출액이 많지 않을때(이상치x)
  - 단기 : 그 외

```python
foreigner=pd.read_csv('foreigner.csv')
foreigner=foreigner.iloc[:,1:]
foreigner=foreigner[foreigner['GB3']!='의료']
foreigner['VLM/USEC']=foreigner['VLM']/foreigner['USEC']
foreigner
```

```python
Q1 = np.percentile(foreigner['USEC'], 25) 
Q3 = np.percentile(foreigner['USEC'], 75) 
IQR = Q3 - Q1 
outlier_step = 1.5 * IQR 
foreigner_long=foreigner[(foreigner['USEC']>Q3+1.5*IQR)]

Q1 = np.percentile(foreigner_long['VLM/USEC'], 25) 
Q3 = np.percentile(foreigner_long['VLM/USEC'], 75) 
IQR = Q3 - Q1 
outlier_step = 1.5 * IQR 
foreigner_long=foreigner_long[(foreigner_long['VLM/USEC']<=Q3+1.5*IQR)]
foreigner_long

Q1 = np.percentile(foreigner_long['USEC'], 25) 
Q3 = np.percentile(foreigner_long['USEC'], 75) 
IQR = Q3 - Q1 
outlier_step = 1.5 * IQR 
foreigner_long=foreigner_long[(foreigner_long['USEC']<=Q3+1.5*IQR)]
foreigner_long

Q1 = np.percentile(foreigner_long['VLM'], 25) 
Q3 = np.percentile(foreigner_long['VLM'], 75) 
IQR = Q3 - Q1 
outlier_step = 1.5 * IQR 
foreigner_long=foreigner_long[(foreigner_long['VLM']<=Q3+1.5*IQR)]
foreigner_long
```
