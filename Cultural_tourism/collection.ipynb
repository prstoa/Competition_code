```python
import pandas as pd
import numpy as np
```

```python
trip=pd.read_csv('trip.csv')
culture=pd.read_csv('culture_performance.csv')
w=pd.read_csv('w.csv')
```

```python
trip=trip.drop(['g_for_inpop','g_kor_outpop','g_income','g_expense'],1)
culture=culture.drop(['Unnamed: 0','g_num_performance','g_num_peropen','g_num_perscreen','g_revenue','g_soldticket'],1)
```

```python
trip2=trip.iloc[:,1:]
culture2=culture.iloc[:,1:]
w2=w.iloc[:,1:]
```

```python
display(trip2,culture2,w2)
```

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
trip2=pd.DataFrame(sd.fit_transform(trip2))
culture2=pd.DataFrame(sd.fit_transform(culture2))
w2=pd.DataFrame(sd.fit_transform(w2))
trip2['date']=trip['date']
culture2['date']=culture['date']
w2['date']=w['date']
```

```python
w2.columns=['w_departstore', 'w_largemarget', 'w_constore', 'w_dutyfreeshop',
       'w_fnb', 'w_clothesfiber', 'w_shoeleather', 'w_cosmetic', 'w_train',
       'w_subway', 'w_bus', 'w_taxi', 'w_flight', 'w_generalstay', 'w_hotel',
       'w_motel', 'w_resort', 'w_kfood', 'w_nkfood', 'w_cafeteria', 'w_pub',
       'w_npub', 'w_artcreate', 'w_lib_historicsite', 'w_amusement',
       'w_hairshop', 'w_spa','date']
```

```python
w2.iloc[:,8:13]
```

```python
w2_dutyfree=pd.DataFrame(w2.iloc[:,3])
w2_dutyfree['date']=w['date']

w2_trans=w2.iloc[:,8:13]
w2_trans=w2_trans.drop(['w_taxi'],1)
w2_trans['date']=w['date']

w2_stay=w2.iloc[:,13:17]
w2_stay['date']=w['date']

w2_etc=w2.iloc[:,22:25]
w2_etc['date']=w['date']
```

```python
display(w2_dutyfree,w2_trans,w2_stay,w2_etc)
```

```python
trip2.to_csv('trip2.csv',index=False)
culture2.to_csv('culture2.csv',index=False)
w2_dutyfree.to_csv('w2_dutyfree.csv',index=False)
w2_trans.to_csv('w2_trans.csv',index=False)
w2_stay.to_csv('w2_stay.csv',index=False)
w2_etc.to_csv('w2_etc.csv',index=False)
```
